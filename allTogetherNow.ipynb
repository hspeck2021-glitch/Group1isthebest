```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


def iniPosition(dX):    
    L = 10
    A=1

    x=np.arange(0,L,dX)

    x_pluck=L/2

    y=np.zeros(x.size)

    for i in range(x.size):
        xi = x[i]
        if xi <= x_pluck:
            y[i] = (A / x_pluck) * xi
        else:
            y[i] = A * (L - xi) / (L - x_pluck)
    return x, y

plt.rcParams["animation.html"] = "jshtml" 
plt.rcParams['figure.dpi'] = 100
plt.ioff()
fig, ax = plt.subplots()

# initialize the object being animated
dX = 0.1
x, y = iniPosition(dX)


def animate(frame):
    plt.cla()
    
    # make a small change to the object
    
    
    
    plt.plot(x, y)
    plt.xticks([])
    plt.yticks([])

FuncAnimation(fig, animate, frames=10)
```
