```python
import numpy as np
import matplotlib.pyplot as plt

def vel(a , v, dt):

    for i in range(a.size()):
    
       v[i] += a[i]*dt
    
    return v
```

```python
import numpy as np
import matplotlib.pyplot as plt

def posFinal(yI,v,a, FrameN, dt):
    
    t = FrameN * dt
    yF = np.zeros()
 
    for i in range(a.size()):
    
       yF[i] = yI[i] + v[i]*t + .5*((a[i])**2)*t
        
    return yF
```
