Here are the equations we are using to put this discretized string into motion.

First, the wave equation describes the relationship between the **acceleration (second t-derivative)** of the y-position of one chunk of string ($y_i$), and the **second x-derivative** of the y-position that same chunk of string.

$$\frac{\partial^2 y_i}{\partial t^2} = c^2 \frac{\partial^2 y_i}{\partial x^2}$$

Second, the second-order finite-difference method provides a formula for the **second x-derivative** of the y-position of a chunk of string, in terms of the y-positions of nearby chunks ($y_{i-1}$, $y_i$, and $y_{i+1}$):

$$\frac{\partial^2 y_i}{\partial x^2} = \frac{y_{i-1}-2y_i+y_{i+1}}{h^2}$$

We can combine these equations to connect the the **acceleration** of the y-position of one chunk of string with the y-positions of nearby chunks.

$$\frac{\partial^2 y_i}{\partial t^2} = c^2 \frac{y_{i-1}-2y_i+y_{i+1}}{h^2}$$
